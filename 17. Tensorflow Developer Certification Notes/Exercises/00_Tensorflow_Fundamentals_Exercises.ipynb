## 1. Create a vector, scalar, matrix and tensor with values of your choosing using `tf.constant()`.

```python
scalar = tf.constant(3)
vector = tf.constant([3,4])
matrix = tf.constant(np.random.randint(0, 100, size = 12).reshape(3,4))
tensor = tf.constant(np.random.randint(0, 10, 120).reshape(2,3,4,5))
scalar.shape, vector.shape, matrix.shape, tensor.shape
```

## 2. Find the shape, rank and size of the tensors you created in 1.

```python
def get_tensor_attributes(tensor):
    print(f"The shape of the tensor: {tensor.shape}")
    print(f"The rank of the tensor: {tensor.ndim}")
    print(f"The size of the tensor: {tf.size(tensor)}")

get_tensor_attributes(scalar)
print("#"*80)
get_tensor_attributes(vector)
print("#"*80)
get_tensor_attributes(matrix)
print("#"*80)
get_tensor_attributes(tensor)
```

## 3. Create two tensors containing random values between 0 and 1 with shape `[5, 300]`.

```python
A = tf.random.uniform((5,300), 0, 1)
B = tf.random.normal((5, 300),0, 1)
A.shape, B.shape,tf.reduce_min(A), tf.reduce_max(A)
```

## 4. Multiply the two tensors you created in 3 using matrix multiplication.

```python
tf.matmul(tf.transpose(A), B).shape, tf.matmul(A, tf.transpose(B)).shape # can use tf.reshape also
```

## 5. Multiply the two tensors you created in 3 using dot product.

```python
tf.tensordot(tf.transpose(A), B, axes = 1).shape, tf.tensordot(A, tf.transpose(B), axes = 1).shape
```

## 6. Create a tensor with random values between 0 and 1 with shape [224, 224, 3].

```python
C = tf.random.uniform((224,224,3), 0, 1)
C.shape
```

## 7. Find the min and max values of the tensor you created in 6 along the first axis.

```python
tf.reduce_max(C, axis = 1), tf.reduce_min(C, axis = 1)
```

## 8. Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].

```python
D = tf.random.normal((1,224,224,3))
D.shape, tf.squeeze(D).shape
```
